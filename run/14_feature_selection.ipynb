```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

# Configuration
import os
import sys
rootpath = os.path.dirname(os.getcwd())
sys.path.append(rootpath)

from news import NewsIO, NewsFunc
newsio = NewsIO()
newsfunc = NewsFunc()

import numpy as np
import pandas as pd
from copy import deepcopy
import matplotlib.pyplot as plt
```

```python
## Parameters
TOPN = 1000

## Filenames
fname_data_norm = f'data_w-{TOPN}_norm.pk'
```

```python
cci = newsio.load_cci(start='200501', end='201912')
cci
```

```python
df_norm = newsio.load(fname_object=fname_data_norm, _type='data')

df_norm
```

```python
data_for_corr = deepcopy(pd.merge(cci, df_norm))
data_for_corr = deepcopy(data_for_corr.drop('yearmonth', axis=1))
data_for_corr
```

```python
df_corr = data_for_corr.corr()['cci']
df_corr
```

```python
corr_list = sorted([(var, df_corr[var]) for var in df_corr.keys() if var != 'cci'], key=lambda x:x[1], reverse=True)
# for var, value in corr_list:
#     print(f'{var}: {value:.03f}')
```

```python
fname_correlated_variables = 'correlated_variables.json'
correlated_variable_list = [var for var, corr in corr_list if corr >= 0.8]
print(len(correlated_variable_list))

newsio.save_json(_object=correlated_variable_list, fname_object=fname_correlated_variables, _type='data')
```
